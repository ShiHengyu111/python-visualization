### 图形尺寸和分辨率设置

**第一个指令：`plt.figure`**


- `plt.figure()`的作用是创建一个新的图形，并返回一个`Figure`对象。`plt.figure()`的参数包括：

- figsize`：图形的大小，是一个包含两个数字的元组，分别代表宽度和高度，单位为英寸（1 inch = 2.54cm）。

- `dpi`：图形的分辨率，即每英寸的点数。

**第二个指令：`fig.add_subplot`**


接下来，`fig.add_subplot()`的作用是向图形添加一个子图，并返回一个Axes对象。`fig.add_subplot()`的参数包括：

- `nrows`：子图的行数。

- `ncols`：子图的列数。

- `index`：子图的编号。

最后，使用Axes.scatter()在子图上绘制散点图，并使用`Axes.set_title()`、`Axes.set_xlabel()`和`Axes.set_ylabel()`添加图形标题和标签。最后使用`plt.show()`显示图形。

```python
import matplotlib.pyplot as plt
import numpy as np

# 创建一个新的图形，并设置它的大小和分辨率
fig = plt.figure(figsize=(12, 4), dpi=800)

# 在图形上添加两个子图，设置子图的位置和大小
# 参数：行数，列数，子图编号
ax1 = fig.add_subplot(1, 3, 1)
ax2 = fig.add_subplot(1, 3, 2)
ax3 = fig.add_subplot(1, 3, 3)

# 生成随机数
x = np.random.normal(size=100)
y = np.random.normal(size=100)

# 在子图1上绘制散点图
ax1.scatter(x, y)
# 在子图2上绘制直方图
ax2.hist(x, bins=20)
# 在子图3上绘制箱线图
ax3.boxplot(x)

# 添加标题和标签
ax1.set_title('Random Scatter Plot')
ax1.set_xlabel('X Values')
ax1.set_ylabel('Y Values')

'''
小练习：
自己给子图二和子图三添加标题和标签。
'''











# 显示图形
plt.show()
```

### Figure对象和Axes对象

`Figure`和`Axes`是Matplotlib中的两个重要概念，它们用于绘制图形和处理图形的各种属性。下面是两者的详细区别：


- `Figure`是**整个图形窗口，可以理解为画布（一张A4纸）**，而`Axes`是在`Figure`上的一个个**独立的绘图区，可以理解为图形中的一个个子区域（你往上面贴一个个便利贴）**。

- 一个`Figure`可以包含多个`Axes`，而一个`Axes`只能属于一个`Figure`。

- `Figure`用于**控制图形的整体布局，如图形大小、分辨率等（便利贴之间的关系）**，而`Axes`用于**控制图形中的各种绘图元素，如坐标轴、数据、标签等（每个便利贴上都写了什么）**。

- 在Matplotlib中，绘制图形的过程是先创建`Figure`，再在`Figure`上添加`Axes`，然后在`Axes`上绘制图形元素。

总的来说，`Figure`和`Axes`都是Matplotlib中非常重要的概念，`Figure`用于控制整个图形的布局和属性，`Axes`用于控制图形中的具体绘图元素。理解`Figure`和`Axes`的区别可以更好地理解Matplotlib的绘图过程，也能更好地实现各种复杂的绘图需求。

```python
'''
绘制四个子图2*2排列
'''

import matplotlib.pyplot as plt
import numpy as np

# 创建一个新的图形
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# 绘制第一个子图
x1 = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x1)
axs[0, 0].plot(x1, y1)
axs[0, 0].set_title('Sin Curve')
axs[0, 0].set_xlabel('X Values')
axs[0, 0].set_ylabel('Y Values')
axs[0, 0].text(0.5, 0.5, 'Text on Sin Curve', ha='center', va='center', fontsize=12)

# 绘制第二个子图
x2 = np.linspace(0, 2 * np.pi, 100)
y2 = np.cos(x2)
axs[0, 1].plot(x2, y2)
axs[0, 1].set_title('Cos Curve')
axs[0, 1].set_xlabel('X Values')
axs[0, 1].set_ylabel('Y Values')
axs[0, 1].annotate('Annotation on Cos Curve', xy=(2, 0.5), xytext=(4, 0.8), arrowprops=dict(facecolor='black', shrink=0.05))

# 绘制第三个子图
x3 = np.linspace(-5, 5, 100)
y3 = x3 ** 2
axs[1, 0].plot(x3, y3)
axs[1, 0].set_title('Square Curve')
axs[1, 0].set_xlabel('X Values')
axs[1, 0].set_ylabel('Y Values')
axs[1, 0].legend(['y=x^2'])

# 绘制第四个子图
x4 = np.linspace(-5, 5, 100)
y4 = x4 ** 3
axs[1, 1].plot(x4, y4)
axs[1, 1].set_title('Cubic Curve')
axs[1, 1].set_xlabel('X Values')
axs[1, 1].set_ylabel('Y Values')
axs[1, 1].legend(['y=x^3'])

# 显示图形
plt.show()
```

### `Figure`对象

**`Figure`对象的属性：(属性的意思就是[figure名字].[属性名字]，如figure.add_subplot(1,2,1))**

- `add_subplot`：向Figure添加一个subplot

- `add_axes`：向Figure添加一个Axes对象

- `subplots`：返回Figure上的所有subplot

- `dpi`：返回Figure的分辨率

- `suptitle`：设置Figure的标题

- `savefig`：保存Figure为图片文件

- `facecolor`：返回Figure的背景颜色

- `set_size_inches`：设置Figure的尺寸

```python
import matplotlib.pyplot as plt
import numpy as np
# 创建一个新的 figure
fig = plt.figure()

# 设置 figure 的大小和分辨率
fig.set_size_inches(8, 6)
fig.set_dpi(100)

# 向 figure 添加一个 subplot
ax1 = fig.add_subplot(221)
ax1.set_title('Subplot 1')
ax1.plot(np.random.rand(10))

# 向 figure 添加一个 Axes 对象
ax2 = fig.add_axes([0.5, 0.2, 0.3, 0.3]) # 参数：[left, bottom, width, height]
ax2.set_title('Subplot 2')
ax2.plot(np.random.rand(5))

# 设置标题和前景色
fig.suptitle('My Figure')
fig.set_facecolor('lightgrey')

# 保存 figure 为图片文件
fig.savefig('my_figure.png')
# 显示 figure
plt.show()
```

### `Axes`对象

**`Axes`对象的属性：(属性的意思就是[figure名字].[属性名字]，如ax.set_xlim("x_label"))**

- `xaxis`：返回Axes的x轴对象

- `yaxis`：返回Axes的y轴对象

- `set_xlabel`：设置Axes的x轴标签

- `set_ylabel`：设置Axes的y轴标签

- `set_xlim`：设置Axes的x轴范围

- `set_ylim`：设置Axes的y轴范围

- `set_title`：设置Axes的标题

- `legend`：返回Axes的图例对象

- `spines`：返回Axes的spines对象，Spines 是指 Axes 四个边缘的线条，即 x 轴、y 轴、左侧边框和底部边框。默认情况下，这些边框是实线，颜色为黑色。

- `grid`：返回Axes的网格线对象

- `get_xlim`：返回Axes的x轴范围

- `get_ylim`：返回Axes的y轴范围

- `set_facecolor`：设置Axes的背景颜色

- `annotations`：返回Axes上的注释列表


```python
import matplotlib.pyplot as plt
import numpy as np
# 创建一个新的 figure 和一个 Axes 对象
fig, ax = plt.subplots()

# 设置 Axes 的标题
ax.set_title('My Plot')

# 设置 Axes 的 x 轴标签和 y 轴标签
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')

# 设置 Axes 的 x 轴范围和 y 轴范围
ax.set_xlim(0, 10)
ax.set_ylim(-5, 5)

# 绘制一条曲线
x = np.linspace(0, 10, 100)
y = np.sin(x)
ax.plot(x, y, label='Sin(x)')

# 添加图例
ax.legend() 

# 获取 Axes 的 spines 对象，并设置其样式和颜色
# Spines 是指 Axes 四个边缘的线条，即 x 轴、y 轴、左侧边框和底部边框。默认情况下，这些边框是实线，颜色为黑色。
spines = ax.spines
for spine in spines.values():
    spine.set_linestyle('--')
    spine.set_color('red')
    
# 获取 Axes 的网格线对象，并设置其样式和颜色
grid = ax.grid(linestyle=':', color='grey') #linestyle可以设置为'-'、'--'、':'、'-.'等。

# 设置 Axes 的背景颜色
ax.set_facecolor('whitesmoke')

# 添加一些注释
ax.annotate('Peak', xy=(np.pi/2, 1), xytext=(4, 2),
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.annotate('Valley', xy=(3*np.pi/2, -1), xytext=(6, -2),
            arrowprops=dict(facecolor='black', shrink=0.05))

# 显示图形
plt.show()
```

**`ax.annotate(text, xy, xytext=None, arrowprops=None, **kwargs)`**

- `text`：要添加的文本内容，可以是字符串或任何可转换为字符串的对象。

- `xy`：注释箭头的目标位置，一个二元组 (x, y)。箭头将指向这个位置。

- `xytext`：注释文本的位置，一个二元组 (x, y)。如果不指定，会自动选择一个合适的位置。可以通过设置 textcoords 参数来控制这个位置是以 Axes 坐标系、数据坐标系还是 Figure 坐标系指定的，默认为 Axes 坐标系。

- `arrowprops`：一个字典，用于设置箭头的样式。常用的键包括 arrowstyle（箭头样式，例如 '->' 表示带箭头的直线，'-|>' 表示带箭头的垂直线）、connectionstyle（连接线样式，例如 'arc3' 表示三次贝塞尔曲线）、facecolor（箭头和连接线的填充色）、edgecolor（箭头和连接线的边框色）等等。

- `**kwargs`：其他参数，例如 textcoords、fontsize、color 等，用于控制文本样式。

```python
# 导入必要的库
from matplotlib.pyplot import MultipleLocator  # 用于设置坐标轴的刻度
import numpy as np
import matplotlib.pyplot as plt
# 设置全局字体
plt.rc('font', family='sans-serif')

# 创建一个新的图形并设置大小和dpi
plt.figure(figsize=(8, 7), dpi=100)

# 设置标题
plt.title("Anacomy of a figure", fontsize=20, fontweight='heavy')

# 获取当前轴
ax = plt.gca()

# 设置x轴和y轴的限制范围
ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

# 设置x轴和y轴的主刻度
xmajorLocator = MultipleLocator(1)
ax.xaxis.set_major_locator(xmajorLocator)
ymajorLocator = MultipleLocator(1)
ax.yaxis.set_major_locator(ymajorLocator)

# 设置x轴和y轴的次刻度
xminorLocator = MultipleLocator(0.25)
ax.xaxis.set_minor_locator(xminorLocator)
yminorLocator = MultipleLocator(0.25)
ax.yaxis.set_minor_locator(yminorLocator)

# # 设置x轴的次刻度标签格式
# xminorFormatter = FormatStrFormatter('%0.2f')
# ax.xaxis.set_minor_formatter(xminorFormatter)

# 设置刻度线的长度
ax.tick_params(which='major', length=8)
ax.tick_params(which='minor', length=4)

# 添加x轴和y轴的网格线
ax.xaxis.grid(True, which='major', linestyle=(0, (8, 4))) 
ax.yaxis.grid(True, which='major', linestyle=(0, (8, 4)))

# 设置x轴和y轴的标签
ax.set_xlabel("X axis label", c='blue', fontsize=12)
ax.set_ylabel("Y axis label", c='blue', fontsize=12)

# 绘制图像
# 绘制红色曲线
x = list(np.arange(0.5, 3.3, 0.1))
y1 = [-0.04658*i**3 + 0.5494*i**2 - 1.3151*i + 1.415 for i in x]
ax.plot(x, y1, color='red', alpha=0.7, linewidth=2.3, label='Red Signal')

# 绘制蓝色曲线
x = list(np.arange(0.5, 3.5, 0.1))
y2 = [0.1545*i**3 - 0.799*i**2 + 0.4316*i + 3.744 for i in x]
ax.plot(x, y2, color='blue', alpha=0.7, linewidth=2.3, label='Blue Signal')

# 添加图例
plt.legend()

# 绘制散点图
A = np.random.normal(loc=2, scale=1, size=30)
B = np.random.normal(loc=2, scale=1, size=30)
ax.scatter(A, B, marker='o', color="w", edgecolors='black')

# 添加文本标签
ax.text(2.9, 2.7, "Gird", family="DejaVu Sans", fontsize=11, color='blue', weight='heavy')
ax.text(3.45, 3.45, "Legend", family="DejaVu Sans", fontsize=10, color='blue', weight='heavy')
ax.text(1.5, 3.85, "Title", family="DejaVu Sans", fontsize=11, color='blue', weight='heavy')
ax.text(-0.25, 3.9, "Major tick", family="DejaVu Sans", fontsize=10, color='blue', weight='heavy')
ax.text(-0.25, 3.3, "Minor tick", family="DejaVu Sans", fontsize=10, color='blue', weight='heavy')
ax.text(-0.45, 2.85, "Major tick label", family="DejaVu Sans", fontsize=10, color='blue', weight='heavy')
ax.text(0.12, -0.3, "Minor tick label", family="DejaVu Sans", fontsize=10, color='blue', weight='heavy')
ax.text(3.25, 0.5, "splines", family="DejaVu Sans", fontsize=10, color='blue', weight='heavy')

ax.arrow(3.3,0.41,-0.15,-0.33,ec='blue',head_width=0.04)
ax.arrow(3.6,0.41,0.32,-0.15,ec='blue',head_width=0.04)
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# 加载数据
iris = load_iris()

# 提取花萼长度和宽度作为变量
sepal_length = iris.data[:, 0]
sepal_width = iris.data[:, 1]
target = iris.target

flg, ax = plt.subplots(figsize=(8, 6))

# 绘制散点图
plt.scatter(sepal_length,sepal_width, c=target, cmap="viridis", s=30, alpha=0.8) # c是颜色，s是大小，alpha是透明度

# 添加标题和轴标签
plt.title("Iris dataset")
plt.xlabel("sepal length(cm)")
plt.ylabel("sepal width(cm)")

# 设置横纵轴刻度
ax.set_xticks(np.arange(4, 8 + 1, 1))
ax.set_yticks(np.arange(2, 5 + 1, 1))

# 设置横纵轴尺度1:1
ax.axis("scaled") # 保持图形在绘制时不会因为坐标轴的比例问题而产生形变

# 增加刻度网络，颜色为浅灰色
plt.grid(linestyle="--", linewidth=0.25, color="lightgray") #color=[0.7,0.7,0.7]

# 设置横纵轴范围
ax.set_xbound(lower = 4, upper = 8)
ax.set_ybound(lower = 1, upper = 5)

# 显示图形 
plt.show()
```

### 直方图

#### 使用matplotlib

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np

data = load_iris().data[ : , 0] # 提取花萼长度数据，即第一列数据

# 创建一个新的 figure 和一个 Axes 对象
fig, ax = plt.subplots(figsize=(8, 6))

# 绘制直方图
n, bins, patches = ax.hist(data, bins=30, density=True, alpha=0.7,
                                color='mediumslateblue', edgecolor='white',
                                label='Data')

# 添加标题和标签
ax.set_title('iris_data', fontsize=20, fontweight='bold', pad=20)
ax.set_xlabel('sepal_length', fontsize=16, labelpad=15) # 设置横轴标签, labelpad设置标签与轴的距离
ax.set_ylabel('Frequency', fontsize=16, labelpad=15)

# 调整坐标轴范围和刻度
ax.set_xlim(4, 8); ax.set_ylim(0, 1)
ax.set_xticks(np.arange(4, 8, 1)); ax.set_yticks(np.arange(0, 1, 0.2))
ax.tick_params(axis='both', labelsize=14) # 设置刻度标签的大小

# 添加图例和网格线
ax.legend(fontsize=14, frameon=False, loc='upper left') 
ax.grid(axis='y', linestyle='--', alpha=0.5) # 添加水平网格线

# 添加注释和数据标签
ax.text(8, 1, 'Mean: {:.2f}\nStd: {:.2f}'.format(data.mean(), data.std()),
        fontsize=14, bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.5))

# 调整布局并显示图像
fig.tight_layout() # 调整布局，使得图像填充整个图像区域
plt.show()
```

### 饼图

```python
import matplotlib.pyplot as plt

# 数据
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]  # 每个部分的大小
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']  # 颜色
explode = (0, 0.1, 0, 0)  # 突出显示某个部分

plt.subplots(figsize=(8, 8))

# 绘制饼图
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140, textprops={'fontsize': 12})

# 添加图例
plt.legend(labels, loc="best")

# 标题
plt.title('Pie Chart Example', fontsize=16, color= 'black')

# 设置边框颜色
plt.gca().spines['top'].set_color('none')
plt.gca().spines['bottom'].set_color('none')
plt.gca().spines['left'].set_color('gray')
plt.gca().spines['right'].set_color('none')
plt.gca().tick_params(axis='both', which='both', length=0)

# 等比例显示，确保饼图是圆的
plt.axis('equal')

# 显示图形
plt.show()
```

### 等高线图

#### 平面等高线

```python
# 为了绘制平面等高线，我们需要利用 numpy.meshgrid() 产生网格数据。
import numpy as np
import matplotlib.pyplot as plt

# 生成数据
x1_array = np.linspace(-3, 3, 121)
x2_array = np.linspace(-3, 3, 121)

xx1, xx2 = np.meshgrid(x1_array, x2_array)
ff = xx1 * np.exp(-xx1**2 - xx2**2)

# 等高线
fig, ax = plt.subplots()

CS = ax.contour(xx1, xx2, ff, levels = 20, cmap="RdYlBu", linewidths = 1) # levels=20 是指定等高线的数量。cmap='RdYlBu_r'是指定等高线颜色映射的参数。例子中使用了 'RdYlBu_r'，表示红、黄、 蓝渐变的颜色映射，_r 表示翻转。
# 这句返回值 CS 是一个等高线图的对象。这个对象包含了等高线图的信息，比如线条的位置、颜色等。

fig.colorbar(CS, ax=ax, shrink=0.8) # 添加颜色条，shrink=0.8 是指定颜色条的长度占整个图形区域宽度的80%。
ax.set_xlabel("x1"); ax.set_ylabel("x2")
ax.set_xticks([]); ax.set_yticks([])
ax.set_xlim(xx1.min(), xx1.max())
ax.set_ylim(xx2.min(), xx2.max())
ax.grid(False)
ax.set_aspect("equal", adjustable="box") # 设置横纵轴比例相同，adjustable="box"表示调整图形的大小以保持横纵轴比例不变。

# 填充等高线(contourf, 比contour多一个f)
fig, ax = plt.subplots()
CS = ax.contourf(xx1, xx2, ff, levels = 20, cmap="RdYlBu") # levels=20 是指定等高线的数量。cmap='RdYlBu_r'是指定等高线颜色映射的参数。例子中使用了 'RdYlBu_r'，表示红、黄、 蓝渐变的颜色映射，_r 表示翻转。

fig.colorbar(CS) # 添加颜色条，shrink=0.8 是指定颜色条的长度占整个图形区域宽度的80%。
ax.set_xlabel("x1"); ax.set_ylabel("x2")
ax.set_xticks([]); ax.set_yticks([])
ax.set_xlim(xx1.min(), xx1.max())
ax.set_ylim(xx2.min(), xx2.max())
ax.grid(False)
ax.set_aspect("equal", adjustable="box") # 设置横纵轴比例相同，adjustable="box"表示调整图形的大小以保持横纵轴比例不变。
```

#### 用plotly绘制二维等高线

```python
import numpy as np
import plotly.graph_objects as go

# 生成数据
x1_array = np.linspace(-3, 3, 121)
x2_array = np.linspace(-3, 3, 121)

xx1, xx2 = np.meshgrid(x1_array, x2_array)
ff = xx1 * np.exp(-xx1**2 - xx2**2)

# 等高线设置
levels = dict(start = -0.5, end = 0.5, size = 0.05) # 设置等高线的起始值、终止值和间隔值
data = go.Contour(z = ff, x = x1_array, y = x2_array, contours_coloring = "fill", 
                    line_width = 2, colorscale = "RdYlBu_r", contours = levels) # contours_coloring="lines"表示等高线的颜色由线条的颜色决定,"fill"会生成填充等高线。line_width=2 表示等高线的线宽为2。

# 图像布局
layout = go.Layout(xaxis_title="x1", yaxis_title="x2", xaxis_showgrid=False, yaxis_showgrid=False, xaxis_ticks="", yaxis_ticks="",
                    xaxis_range=[-3, 3], yaxis_range=[-3, 3], width=600, height=600)

# 创建图形对象
fig = go.Figure(data = data, layout = layout)
fig.show()
```

### 三维线图

#### 使用matplotlib

```python
# 导入包
import matplotlib.pyplot as plt
import numpy as np
import plotly.graph_objects as go

# 生成随机游走数据
num_steps = 300
t = np.arange(num_steps)
x = np.cumsum(np.random.standard_normal(num_steps))
y = np.cumsum(np.random.standard_normal(num_steps))
z = np.cumsum(np.random.standard_normal(num_steps))

# 用 Matplotlib 可视化
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(x,y,z,color = 'darkblue')
ax.scatter(x,y,z,c = t, cmap = 'viridis')

ax.set_xticks([]); ax.set_yticks([]); ax.set_zticks([])
# 设置正交投影
ax.set_proj_type('ortho')
# 设置相机视角
ax.view_init(elev = 30, azim = 120)
# 显示图像
plt.show()
```

#### 使用plotly

```python
# 用 Plotly 可视化
fig = go.Figure(data=go.Scatter3d(
    x=x, y=y, z=z,
    marker=dict(size=4,color=t,colorscale='Viridis'),
    line=dict(color='darkblue', width=2)))

fig.layout.scene.camera.projection.type = "orthographic"
fig.update_layout(width=800,height=700)
fig.show()  # 显示绘图结果
```

### 箭头图（二维/三维）

```python
# 导入包
import matplotlib.pyplot as plt

# 定义二维列表
A = [[0,5], [3,4], [5,0]]

# 自定义可视化函数
def draw_vector(vector,RBG): 
    plt.quiver(0, 0, vector[0], vector[1],angles='xy', 
                scale_units='xy',scale=1,color = RBG,
                zorder = 1e5)

fig, ax = plt.subplots()
v1 = A[0] # 第一行向量
draw_vector(v1,'#FFC000')
v2 = A[1] # 第二行向量
draw_vector(v2,'#00CC00')
v3 = A[2] # 第三行向量
draw_vector(v3,'#33A8FF')

ax.axvline(x = 0, c = 'k')
ax.axhline(y = 0, c = 'k')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.grid()
ax.set_aspect('equal', adjustable='box')
ax.set_xbound(lower = -0.5, upper = 5)
ax.set_ybound(lower = -0.5, upper = 5)


# 自定义可视化函数
def draw_vector_3D(vector,RBG): 
    plt.quiver(0, 0, 0, vector[0], vector[1], vector[2],
                arrow_length_ratio=0, color = RBG,
                zorder = 1e5)
    
fig = plt.figure(figsize = (6,6))
ax = fig.add_subplot(111, projection='3d', proj_type = 'ortho')
# 第一列向量
v_1 = [row[0] for row in A]
draw_vector_3D(v_1,'#FF6600')
# 第二列向量
v_2 = [row[1] for row in A]
draw_vector_3D(v_2,'#FFBBFF')

ax.set_xlim(0,5)
ax.set_ylim(0,5)
ax.set_zlim(0,5)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('x3')
ax.view_init(azim = 30, elev = 25)
ax.set_box_aspect([1,1,1])
```
